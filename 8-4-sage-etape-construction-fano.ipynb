```python
%matplotlib inline
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
 
def draw_plane_with_3points(points):
 
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
 
    p_array = np.array(points)
    p_array_T = p_array.T
    x = p_array_T[0]
    y = p_array_T[1]
    z = p_array_T[2]
 
    X, Y = np.meshgrid(x, y)
    _, Z = np.meshgrid(z, z)
    # Z = z.dot(np.ones([3,3])) # use z.T
 
    ax.plot_surface(X, Y, Z)
 
    ax.set_xlabel('X Label')
    ax.set_ylabel('Y Label')
    ax.set_zlabel('Z Label')
 
    ax.plot(x, y, z, '*')
 
    plt.show()
 
 
points = [
    [1, 1, 1],
    [2, 1, 1],
    [2, 2, 2],
]
 
draw_plane_with_3points(points)
```
